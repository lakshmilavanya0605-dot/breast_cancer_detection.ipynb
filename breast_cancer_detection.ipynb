Breast Cancer Detection using Machine Learning

Step-1: Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Step-2: Data Collection

```python
# Load dataset
df = pd.read_csv("data.csv")

# Show first 5 rows
df.head()
```

Step-4: Data Preprocessing

```python
# Convert diagnosis (M=1, B=0)
df['diagnosis'] = df['diagnosis'].map({'M':1, 'B':0})

# Drop id column
if 'id' in df.columns:
    df = df.drop('id', axis=1)

df.head()
```

Step-5: Splitting Data

```python
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Step-6: Model Training

```python
# Fix data issues before training

# Remove any non-numeric columns
X = X.select_dtypes(include=['number'])

# Fill missing values if any
X = X.fillna(0)

# Split again
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=10000)

model.fit(X_train, y_train)
```

Step-7: Model Evaluation

```python
# Predict values for test data
y_pred = model.predict(X_test)

# Import accuracy score
from sklearn.metrics import accuracy_score

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Display result
print("Accuracy:", accuracy)
```

Step-8: Prediction System

```python
# Take one sample from test data
sample = X_test.iloc[0].values.reshape(1, -1)

# Make prediction
prediction = model.predict(sample)

# Show result
if prediction[0] == 1:
    print("The tumor is Malignant (Cancer)")
else:
    print("The tumor is Benign (No Cancer)")
```

Step-9: Conclusion


In this project, we built a Machine Learning model to detect breast cancer using logistic regression.
The model was trained on medical data and achieved good accuracy in predicting whether a tumor is malignant or benign.
This system can help in early detection of breast cancer and support medical diagnosis.
